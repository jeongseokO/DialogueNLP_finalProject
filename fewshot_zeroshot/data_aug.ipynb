# [INST][/INST]를 모든 대화에 적용

```python

import numpy as numpy
import pandas as pd
import json

inputs = []
outputs = []
data = []
def add_tags(text, start_tag="[INST]", end_tag="[/INST]"):
    # 분할된 각 부분에 태그 추가
    tagged_parts = [start_tag + part + end_tag for part in text.split("<")[1:]]
    # 태그가 추가된 문자열을 다시 합침
    #tagged_text = "<user>".join(tagged_parts)
    return tagged_parts
with open("../../../data/jeongseokoh/dataset/inform-multi+multi/test-clean.jsonlist", 'r', encoding='utf-8') as file:
    for line in file:
        dialog = json.loads(line)
        data.append(dialog)

for dialog in data:
    input = dialog['input']
    input = add_tags(input)
    #print(input)
    attached_part = ""
    for part in input:
        
        for word in part.split("[INST]")[1:]:
            #print('[INST]'+'<user>'+word)
            attached_part += '[INST]'+'<'+word
        #print(attached_part.replace("\n",""))
    output = dialog['output']
    inputs.append(attached_part)
    outputs.append(output)
    

data = {"input":inputs, "output":outputs}

#with open("intent_test_data.json","w") as file:
#    json.dump(data, file)
for i in range(20):
    input = data['input'][2000+i]
    output = data['output'][2000+i]
    print(f'input: {input}')
    print(f'output: {output}')
```
